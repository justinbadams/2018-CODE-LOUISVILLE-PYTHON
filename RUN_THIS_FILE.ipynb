```python
import csv, sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')

# load/read data from data.csv into pandas dataframe 
df = pd.read_csv('alldata.csv', encoding='utf-8', low_memory=False)
df # print out data to verify it's there
```

```python
# make connection to sqlite database and create new db or replace if exists
conn = sqlite3.connect('all_sales.db')

# creates sql table
df.to_sql('all_sales_table', conn, if_exists='replace')

# view chosen columns and alias from table and organize by description
df_all_sales = pd.read_sql_query('SELECT ordered_item_style AS Description, ordered_item_pattern AS Pattern, coupon_code AS Coupon FROM all_sales_table ORDER BY ordered_item_style', conn)
df_all_sales
```

```python
# new data set to count concatenated description (concat: collection || print_pattern)
df_pattern_sales = pd.read_sql_query('SELECT ordered_item_style AS Collection, ordered_item_pattern AS Print_Pattern, COUNT(*) AS Items_Sold, ordered_item_style || "_" || ordered_item_pattern AS Desc FROM all_sales_table GROUP BY ordered_item_style, ordered_item_pattern ORDER BY Items_Sold DESC', conn)
df_pattern_sales
```

```python
# creates plots
f, ax = plt.subplots(figsize=(10,20))

# bar chart
sns.barplot(x='Collection', y='Description', data=df_pattern_sales, color='m', label='Total Sales by Each Size and Pattern')

# label the indexes
ax.set_xlabel('Total Items Sold')
ax.xaxis.set_label_position('top') 
ax.set_ylabel('Style-Size-Pattern') 
ax.xaxis.tick_top()
ax.legend(loc='center right')
plt.show()

```
